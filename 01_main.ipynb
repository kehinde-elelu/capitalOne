```python
import pandas as pd
import numpy as np
import zipfile
import os
```

```python
#Unzip the downloaded file into a data folder name 
baseUrl = '/Users/kehindeelelu/Documents/Code/CapitalOne'
def unzip_file(zip_file, extract_to):
    with zipfile.ZipFile(zip_file, 'r') as zip_ref:
        zip_ref.extractall(extract_to)

# Specify the path to your zip file and where you want to extract it
zip_file = f'{baseUrl}/transactions.zip'
extract_to = f'{baseUrl}/data'

unzip_file(zip_file, extract_to)
```

```python
df = pd.read_json('/Users/kehindeelelu/Documents/Code/CapitalOne/data/transactions.txt', lines=True)
# Display the first few rows of the DataFrame
df.head()
```

```python
# Check if all columns contain empty strings
all_empty_strings = df.apply(lambda col: col.astype(str).str.strip().eq('')).all()
```

```python
# Print the result (showing first 15 columns separately for readability)
print("All columns contain empty strings (First 15 columns):")
print(all_empty_strings[:15])
```

```python
print("All columns contain empty strings (From 16th column onwards):")
print(all_empty_strings[15:])
```

```python
# Replace all empty strings with NAN
df.replace('', np.nan, inplace=True)
```
